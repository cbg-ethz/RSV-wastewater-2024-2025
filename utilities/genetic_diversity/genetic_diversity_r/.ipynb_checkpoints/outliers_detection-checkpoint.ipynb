```r
# OUTLIERS IDENTIFICATION following McLeod R. et al, 2025, preprint "Real-time outlier detection in digital PCR..."
 
library(ggplot2)
library(dplyr)
library(ggplot2)
library(robustbase)
library(stringr)
library(tidyr)
library(slider)
library(matrixStats)

library(lubridate)
library(patchwork)
library(purrr)
```

```r
# ──────────────────────────────────────────────────────────────────────────
# Parameters
# ──────────────────────────────────────────────────────────────────────────

lod <- 0.006# 6 gc/ml wastewater
# d = dilution factor of the sample
d = 30
# n = replicates per sample
n = 2
# p = average number of partitions in the rxn
p = 22000
# rolling-median window size
window_size <- 13
v <- 5.19e-4
z_threshold <- 3



# ------------------------ List of the dates at which the samples were sequenced for each treatment plant ------------------------
samples <- read.csv("/Users/arimaite/Documents/GitHub/combined_virus_sequencing/RSV/data_analysis/data/samples_tsv/samples_corrected.tsv",
                    sep = '\t',
                    header = TRUE,
                    colClasses = c(location = "character"))
samples$date <- as.character(samples$date)

samples$location <- gsub("10", "Zurich", samples$location)
samples$location <- gsub("16", "Geneva", samples$location)
samples$location <- gsub("15", "Basel", samples$location)
samples$location <- gsub("05", "Lugano", samples$location)
samples$location <- gsub("25", "Laupen", samples$location)
samples$location <- gsub("17", "Chur", samples$location)

samples <- samples %>%
  mutate(SampleName=str_extract(ID, "\\d{2}_\\d{4}_\\d{2}_\\d{2}"))
samples$date <- gsub("_", "-", samples$date)

# Have a list of samples that were sequenced for RSV-A:
rsva_samples <- samples %>%
  filter(!grepl("B", samples$protocol)) %>%
  mutate(SampleName=str_extract(ID, "\\d{2}_\\d{4}_\\d{2}_\\d{2}"))
# Have a list of samples that were sequenced for RSV-B:
rsvb_samples <- samples %>%
  filter(!grepl("A", samples$protocol)) %>%
  mutate(SampleName=str_extract(ID, "\\d{2}_\\d{4}_\\d{2}_\\d{2}"))
```

```r
# ------------------------ Concentration RSV-AB assay ------------------------

# Get the loads from file
rsv_ab_assay <- read.csv(file = '/Users/arimaite/Documents/GitHub/combined_virus_sequencing/RSV/data_analysis/data/concentrations/rsv_ab_assay/251023_RSVAB.csv',
                         check.names = FALSE) %>%
  select(-c(1)) %>%
  rename(Date = collection_date) %>%
  rename(location = ara_name) %>%
  select(Date, sample_id, pathogen, location, conc, flow)  %>%
  #conc (gc/lww), flow (m^3/day)

  pivot_wider(
    names_from = pathogen,
    values_from = conc
  ) %>%
  rename(
    "conc_RSVA" = `RSV-A`,
    "conc_RSVB" = `RSV-B`
  ) %>%
  mutate(
    flow = flow * 1e9, # m^3 per day --> ml per day
    conc_RSVA = conc_RSVA * 1e9, # gc/l --> gc/ml
    conc_RSVB = conc_RSVB * 1e9 # gc/l --> gc/ml

  )



rsv_ab_assay$Date <- gsub("_", "-", rsv_ab_assay$Date)


rsv_ab_assay$location <- gsub("Werdhölzli", "Zurich", rsv_ab_assay$location)
rsv_ab_assay$location <- gsub("Aire", "Geneva", rsv_ab_assay$location)
rsv_ab_assay$location <- gsub("ARA Basel", "Basel", rsv_ab_assay$location)
rsv_ab_assay$location <- gsub("CDA Lugano", "Lugano", rsv_ab_assay$location)
rsv_ab_assay$location <- gsub("ARA Sensetal", "Laupen", rsv_ab_assay$location)
rsv_ab_assay$location <- gsub("ARA Chur", "Chur", rsv_ab_assay$location)

rsv_ab_assay$Date <- as.Date(rsv_ab_assay$Date,"%Y-%m-%d")
rsv_ab_assay <- rsv_ab_assay[rsv_ab_assay$location %in% c("Zurich", "Geneva", "Basel", "Lugano", "Laupen", "Chur"), ]



for (i in 1:nrow(samples)){
  date = samples[i, "date"]
  loc = samples[i, "location"]
  sample_id <- samples[i, "SampleName"]


  rsv_ab_assay[(rsv_ab_assay$Date == date & rsv_ab_assay$location==loc), 'ID'] <- sample_id
}

# Filter samples that were not sequenced
rsv_ab_assay <- rsv_ab_assay %>%
  filter(!is.na(ID))
```

```r
# ──────────────────────────────────────────────────────────────────────────
# # Select pre-PCR coefficient of variation for RSV-AB subtyping assay
# ──────────────────────────────────────────────────────────────────────────
df <- rsv_ab_assay %>% mutate(is_outlier_RSV = FALSE)
# Step 1. Fit trend line
df <- df %>% mutate(is_outlier_RSVA = FALSE, is_outlier_RSVB = FALSE)
# Step 1. Fit trend line
# Calculate normalized concentrations conc_norm = conc * flow
df <- df %>%
  mutate(
    conc_norm_RSVA = conc_RSVA * flow,
    conc_norm_RSVB = conc_RSVB * flow
    
  )

df <- df %>%
  arrange(location, Date) %>%
  group_by(location) %>%
  mutate(
    conc_norm_RSVA_WRM = slide_index2_dbl(
      .x = conc_norm_RSVA, 
      .y = Date, 
      .i = Date,
      .before = days(window_size),
      .after  = days(0),
      .f = ~ {
        center <- .y[length(.y)]
        
        dist_days <- abs(as.numeric(.y - center))
        w <- max(dist_days) - dist_days + 1
        w <- w / sum(w)
        # weighted median
        weightedMedian(.x, w = w, na.rm = TRUE)
      },
      .complete = FALSE
    ),
     conc_norm_RSVB_WRM = slide_index2_dbl(
      .x = conc_norm_RSVB, 
      .y = Date,
      .i = Date,
      .before = days(window_size),
      .after  = days(0),
      .f = ~ {
        center <- .y[length(.y)]
        
        dist_days <- abs(as.numeric(.y - center))
        w <- max(dist_days) - dist_days + 1
        w <- w / sum(w)
        # weighted median
        weightedMedian(.x, w = w, na.rm = TRUE)
      },
      .complete = FALSE
    )
  ) %>% ungroup()



# Step 2. Denormalize the trend
df <- df %>%
  mutate(
    conc_denorm_RSVA_WRM = pmax(conc_norm_RSVA_WRM / flow, lod),
    conc_denorm_RSVB_WRM = pmax(conc_norm_RSVB_WRM / flow, lod)
    
  )

results <- data.frame(
  niu = numeric(),
  var_delta_c_RSVA  = numeric(), 
  abs_dist_to_one_RSVA  = numeric(),
  var_delta_c_RSVB  = numeric(), 
  abs_dist_to_one_RSVB  = numeric()
)
for (niu_unknown in seq(0, 100, by = 0.1)) {
  
  df_tmp <- df %>%
    mutate(
      cv_RSVA = sqrt(
        niu_unknown^2 * 1 / ((conc_denorm_RSVA_WRM / d) * p * n) +
          (1 + niu_unknown^2) / (2 * p * n)
      ),
      
      delta_c_RSVA = (conc_RSVA - conc_denorm_RSVA_WRM) /
        (cv_RSVA * conc_denorm_RSVA_WRM),
      
      cv_RSVB = sqrt(
        niu_unknown^2 * 1 / ((conc_denorm_RSVB_WRM / d) * p * n) +
          (1 + niu_unknown^2) / (2 * p * n)
      ),
      
      delta_c_RSVB = (conc_RSVB - conc_denorm_RSVB_WRM) /
        (cv_RSVB * conc_denorm_RSVB_WRM)
    )
  
  results <- rbind(
    results,
    data.frame(
      niu = niu_unknown,
      var_delta_c_RSVA = var(df_tmp$delta_c_RSVA, na.rm = TRUE),
      abs_dist_to_one_RSVA = abs(1 - var(df_tmp$delta_c_RSVA, na.rm = TRUE)),
      var_delta_c_RSVB = var(df_tmp$delta_c_RSVB, na.rm = TRUE),
      abs_dist_to_one_RSVB = abs(1 - var(df_tmp$delta_c_RSVB, na.rm = TRUE))
      
    )
  )
}

niu_RSVA  <- results[results$abs_dist_to_one_RSVA == min(results$abs_dist_to_one_RSVA), "niu"]
niu_RSVB  <- results[results$abs_dist_to_one_RSVB == min(results$abs_dist_to_one_RSVB), "niu"]

print(niu_RSVA)
print(niu_RSVB)
```

```r
# ──────────────────────────────────────────────────────────────────────────
# # Method to identify the outliers
# ──────────────────────────────────────────────────────────────────────────
identify_outliers <- function(df){
  df <- df %>% mutate(is_outlier_RSV = FALSE)
  df_return <- df %>% mutate(is_outlier_RSV = FALSE)

outlier_df <- data.frame()
    # remove outliers in iterations
    for (iter in 1:10){
  # Remove already identified outliers from the dataset
  # Step 1. Fit trend line
  # Calculate normalized concentrations conc_norm = conc * flow
  
  if (iter > 1) {outlier_df <- rbind(outlier_df, df[df$is_outlier_RSV == TRUE, ])}
  else if (iter == 1) {outlier_df <- df[df$is_outlier_RSV == TRUE, ]}
        
  df <- df[df$is_outlier_RSV == FALSE, ] %>%
    mutate(
      conc_norm_RSV = conc_RSV * flow,

    )
  df <- df %>%
    arrange(location, Date) %>%
    group_by(location) %>%
    mutate(
      conc_norm_RSV_WRM = slide_index2_dbl(
        .x = conc_norm_RSV, .y = Date, .i = Date,
        .before = days(window_size),
        .after  = days(window_size),
        .f = ~ {
          center <- .y[length(.y)]
          
          dist_days <- abs(as.numeric(.y - center))
          w <- max(dist_days) - dist_days + 1
          w <- w / sum(w)
          # weighted median
          weightedMedian(.x, w = w, na.rm = TRUE)
        },
        .complete = FALSE
      )
    ) %>% ungroup()




  # Step 2. Denormalize the trend
  df <- df %>%
    mutate(
      conc_denorm_RSV_WRM = pmax(conc_norm_RSV_WRM / flow, lod),

    )
  

  # Step 3. Predict the coefficient of variation for non-outlier dPCR measurements on each date
  df <- df  %>%

    mutate(

      cv_RSV = sqrt(niu^2 * 1/ ((conc_denorm_RSV_WRM / d) * p * n ) + (1 + niu^2)/(2 * p * n)) 
      
    )

  # Step 4. Standardise deviations
  df <- df %>%
    mutate(
      delta_c_RSV = (conc_RSV - conc_denorm_RSV_WRM) / (cv_RSV * conc_denorm_RSV_WRM)

    )

  # Step 5. Apply z-score threshold
  df <- df %>%
    mutate(
      is_outlier_RSV = ifelse(delta_c_RSV > z_threshold, 1, 0),

    )
  print(sum(df$is_outlier_RSVA, na.rm = TRUE))
  if (sum(df$is_outlier_RSV, na.rm = TRUE) == 0) break

    }

  return (outlier_df)

}
```

```r
# ============================================================================
# ============================================================================
# Outliers in RSV-AB subtyping assay
# ============================================================================

identify_outliers <- function(df){
  dfA <- df %>%
    mutate(is_outlier_RSVA = FALSE)
  dfB <- df %>%
    mutate(is_outlier_RSVB = FALSE)

    outlier_df_RSVA <- data.frame()
    outlier_df_RSVB <- data.frame()

    # remove outliers in iterations
    for (iter in 1:10){
  # Remove already identified outliers from the dataset
  # Step 1. Fit trend line
  # Calculate normalized concentrations conc_norm = conc * flow
  
  if (iter > 1) {
      outlier_df_RSVA <- rbind(outlier_df_RSVA, dfA[dfA$is_outlier_RSVA == TRUE, ])
      outlier_df_RSVB <- rbind(outlier_df_RSVB, dfB[dfB$is_outlier_RSVB == TRUE, ])
  }
        
  else if (iter == 1) {
      outlier_df_RSVA <- dfA[dfA$is_outlier_RSVA == TRUE, ]
      outlier_df_RSVB <- dfB[dfB$is_outlier_RSVB == TRUE, ]
  }
        
  dfA <- dfA[dfA$is_outlier_RSVA == FALSE, ] %>%
    mutate(
      conc_norm_RSVA = conc_RSVA * flow,

    )
  dfB <- dfB[dfB$is_outlier_RSVB == FALSE, ] %>%
    mutate(
      conc_norm_RSVB = conc_RSVB * flow,

    )

        
  dfA <- dfA %>%
    arrange(location, Date) %>%
    group_by(location) %>%
    mutate(
      conc_norm_RSVA_WRM = slide_index2_dbl(
        .x = conc_norm_RSVA, .y = Date, .i = Date,
        .before = days(window_size),
        .after  = days(0),
        .f = ~ {
          center <- .y[length(.y)]
          
          dist_days <- abs(as.numeric(.y - center))
          w <- max(dist_days) - dist_days + 1
          w <- w / sum(w)
          # weighted median
          weightedMedian(.x, w = w, na.rm = TRUE)
        },
        .complete = FALSE
      )) %>% ungroup()

        
  dfB <- dfB %>%
    arrange(location, Date) %>%
    group_by(location) %>%
    mutate(        
      conc_norm_RSVB_WRM = slide_index2_dbl(
        .x = conc_norm_RSVB, .y = Date, .i = Date,
        .before = days(window_size),
        .after  = days(0),
        .f = ~ {
          center <- .y[length(.y)]
          
          dist_days <- abs(as.numeric(.y - center))
          w <- max(dist_days) - dist_days + 1
          w <- w / sum(w)
          # weighted median
          weightedMedian(.x, w = w, na.rm = TRUE)
        },
        .complete = FALSE
      )
    ) %>% ungroup()





  # Step 2. Denormalize the trend
  dfA <- dfA %>%
    mutate(
      conc_denorm_RSVA_WRM = pmax(conc_norm_RSVA_WRM / (flow), lod),

    )
  dfB <- dfB %>%
    mutate(
      conc_denorm_RSVB_WRM = pmax(conc_norm_RSVB_WRM / (flow), lod),

    )

  # Step 3. Predict the coefficient of variation for non-outlier dPCR measurements on each date

  dfA <- dfA %>%
    mutate( # should be gc /ml
      conc_denorm_RSVA_WRM = conc_denorm_RSVA_WRM,
    ) %>%
    mutate(
      cv_RSVA = sqrt(niu_RSVA^2 * 1/ ((conc_denorm_RSVA_WRM / d) * p * n ) +
               (1 + niu_RSVA^2)/(2 * p * n)),
          
        
       #cv_RSVA = sqrt(exp(conc_denorm_RSVA_WRM / d * v ) - 1) / (sqrt(p * n) * conc_denorm_RSVA_WRM/ d * v )
    )

  dfB <- dfB %>%
    mutate( # should be gc /ml
      conc_denorm_RSVB_WRM = conc_denorm_RSVB_WRM
    ) %>%
    mutate(
      cv_RSVB =  sqrt(niu_RSVB^2 * 1/ ((conc_denorm_RSVB_WRM / d) * p * n ) +
               (1 + niu_RSVB^2)/(2 * p * n))
      #cv_RSVB = sqrt(exp(conc_denorm_RSVB_WRM / d * v ) - 1) / (sqrt(p * n) * conc_denorm_RSVB_WRM/ d * v )

    )

  # Step 4. Standardise deviations

  dfA <- dfA %>%
    mutate(
      delta_c_RSVA = (conc_RSVA - conc_denorm_RSVA_WRM) / (cv_RSVA * conc_denorm_RSVA_WRM)
    )
  dfB <- dfB %>%
    mutate(
      delta_c_RSVB = (conc_RSVB - conc_denorm_RSVB_WRM) / (cv_RSVB * conc_denorm_RSVB_WRM)
    )

  # Step 5. Apply z-score threshold

  dfA <- dfA %>%
    mutate(
      is_outlier_RSVA = ifelse(delta_c_RSVA > z_threshold, 1, 0)

    )
  dfB <- dfB %>%
    mutate(
      is_outlier_RSVB = ifelse(delta_c_RSVB > z_threshold, 1, 0)
    )
    print(paste0("Iteration ", iter))
  print(sum(dfA$is_outlier_RSVA, na.rm = TRUE))
  print(sum(dfB$is_outlier_RSVB, na.rm = TRUE))
  if ((sum(dfA$is_outlier_RSVA, na.rm = TRUE) == 0) & (sum(dfB$is_outlier_RSVB, na.rm = TRUE) == 0)) break

    }
  return (list(RSVA = outlier_df_RSVA, RSVB = outlier_df_RSVB))


}
```

```r
options(repr.plot.width = 20, repr.plot.height = 10)

# ===================== RUN IT =====================
result <- identify_outliers(
  rsv_ab_assay
)
rsv_outA <- result$RSVA
rsv_outB <- result$RSVB

# Final result
rsv_outA %>%
  summarise(
    RSVA_outliers = sum(is_outlier_RSVA, na.rm = TRUE),
  )
rsv_outB %>%
  summarise(
    RSVB_outliers = sum(is_outlier_RSVB, na.rm = TRUE),
  )
rsv_ab_assay <- rsv_ab_assay %>%
  mutate(is_outlier_RSVA = ifelse(sample_id %in% rsv_outA$sample_id, TRUE, FALSE),
         is_outlier_RSVB = ifelse(sample_id %in% rsv_outB$sample_id, TRUE, FALSE))


#options(repr.plot.width = 16, repr.plot.height = 8)
long_df <- rsv_ab_assay %>%
  pivot_longer(
    cols = c(conc_RSVA, conc_RSVB),
    names_to = "type",
    values_to = "Concentration"
  )
long_df <- long_df %>%
  mutate(
    outlier_type = case_when(
      is_outlier_RSVA == 1 & type == "conc_RSVA" ~ "RSVA outlier",
      is_outlier_RSVB == 1 & type == "conc_RSVB" ~ "RSVB outlier",
      TRUE ~ NA_character_
    )
  )
max_conc <- max(long_df$Concentration)

g2_pub <- ggplot(long_df, aes(x = Date, y = Concentration, color = type)) +
  
  # ---- Main points and line ----
  geom_point(size = 2.5, alpha = 0.8) +
  geom_line(linewidth = 1) +
  ylim(c(0, max_conc)) +
  
  # ---- Outlier layer ----
  geom_point(
    data = long_df %>% filter(!is.na(outlier_type)),
    aes(color = outlier_type),
    size = 4,
    shape = 21,
    stroke = 1.5
  ) +
  
  # ---- Facet by location ----
  facet_wrap(~ location, scales = "free_y") +
  
  # ---- Labels ----
  labs(
    x = "Date",
    y = "RSV concentration (gc/ul wastewater)",
    title = "RSVA & RSVB outliers",
    color = "Subtype / Outlier"
  ) +
  
  # ---- Colors ----
  scale_color_manual(
values = c(
  "conc_RSVA"    = "#1f77b4", 
  "conc_RSVB"    = "#2ca02c", 
  "RSVA outlier" = "#e41a1c", 
  "RSVB outlier" = "#ff7f00"   
)
  ) +
  
  # ---- Publication-quality theme ----
  theme_minimal(base_size = 18) +        # increase base font size
  theme(
    plot.title = element_text(face = "bold", size = 24, hjust = 0.5),
    axis.title = element_text(face = "bold", size = 20),
    axis.text = element_text(face = "bold", size = 16, angle = 45, hjust = 1),
    legend.title = element_text(face = "bold", size = 18),
    legend.text = element_text(size = 16),
    panel.grid.major = element_line(color = "grey90", linewidth = 0.5),
    panel.grid.minor = element_blank(),
    panel.border = element_rect(color = "black", fill = NA, linewidth = 1.2),
    strip.text = element_text(face = "bold", size = 20)

  )

print(g2_pub)
```

```r
rsv_div_matrix <- outliers_conc_diversity
options(repr.plot.width = 6, repr.plot.height = 8)

# Subset RSV-A
rsv_div_matrix_A <- rsv_div_matrix %>% 
  filter(subtype == "RSV_A") %>% 
  filter(genome_region == "genome") %>%
  mutate(Date = as.Date(Date))

# Create list of outliers

outlier_list <- rsv_div_matrix_A %>%
  filter(is_outlier_RSVA, genome_region == "genome") %>%
  mutate(outlier_id = row_number()) %>%
  group_split(outlier_id)




# Find neighbors: x days and same location
neighbor_windows <- map(outlier_list, function(out) {

  out_date <- out$Date[1]
  out_loc  <- out$location[1]

  neighbors <- rsv_div_matrix_A %>%
    filter(
      #(abs(as.numeric(Date - out_date)) <= 21),
      (abs(as.numeric(Date - out_date)) != 0),
      location == out_loc                      
    )

  list(
    outlier = out,
    neighbors = neighbors
  )
})

wilcox_single_sample_results <- map_df(neighbor_windows, function(x) {

  outlier_entropy  <- x$outlier$entropy
  neighbor_entropy <- x$neighbors$entropy

  test <- wilcox.test(neighbor_entropy, mu = outlier_entropy)

  tibble(
    outlier_sample = x$outlier$ID.x	,
    outlier_date   = x$outlier$Date,
    location       = x$outlier$location,
    W              = test$statistic,
    p_value        = test$p.value,
    n_neighbors    = length(neighbor_entropy)
  )
})

wilcox_single_sample_results <- wilcox_single_sample_results %>%
  mutate(
    p_adj = p.adjust(p_value, method = "BH"),       # FDR correction
    signif = case_when(
      p_adj <= 0.001 ~ "***",
      p_adj <= 0.01  ~ "**",
      p_adj <= 0.05  ~ "*",
      TRUE           ~ "ns"                         # not significant
    )
  )
wilcox_single_sample_results
```

```r
# ============================================================================
# ============================================================================
# Outliers in pan-RSV assay
# ============================================================================
# ============================================================================

# ------------------------------------------------------------------
# PAN-RSV CONCENTRATIONS: OUTLIER DETECTION (McLeod et al. 2025)
# ------------------------------------------------------------------


# ─── 1. Data loading ─────────────────────────────────────
pan_rsv_assay <- read.csv(file = '/Users/arimaite/Documents/GitHub/combined_virus_sequencing/RSV/data_analysis/data/concentrations/pan_rsv/250725_RESP6_RSV_average_unnormalized_total.csv',
                          check.names = FALSE) %>%
  select(-1) %>%
  rename(Date = collection_date,
         location = wastewater_treatment_plant.name,
         conc_RSV = avg_gc_per_lww) %>%
  select(Date, location, conc_RSV, flow) %>%
  mutate(flow = flow * 1e6,
         conc_RSV = conc_RSV * 1e-3) %>%
  mutate(Date = as.Date(gsub("_", "-", Date)),
         location = recode(location,
                           "ARA Werdhoelzli" = "Zurich",
                           "STEP Aire" = "Geneva",
                           "ARA Basel/Prorheno" = "Basel",
                           "CDA Lugano" = "Lugano",
                           "ARA Sensetal" = "Laupen",
                           "ARA Chur" = "Chur"))


# ──────────────────────────────────────────────────────────────────────────
# # Select pre-PCR coefficient of variation
# ──────────────────────────────────────────────────────────────────────────
df <- pan_rsv_assay %>% mutate(is_outlier_RSV = FALSE)
# Step 1. Fit trend line
# Calculate normalized concentrations conc_norm = conc * flow
df <- df %>%
  mutate(
    conc_norm_RSV = conc_RSV * flow,
    
  )

df <- df %>%
  arrange(location, Date) %>%
  group_by(location) %>%
  mutate(
    conc_norm_RSV_WRM = slide_index2_dbl(
      .x = conc_norm_RSV, .y = Date, .i = Date,
      .before = days(window_size),
      .after  = days(window_size),
      .f = ~ {
        center <- .y[length(.y)]
        
        dist_days <- abs(as.numeric(.y - center))
        w <- max(dist_days) - dist_days + 1
        w <- w / sum(w)
        # weighted median
        weightedMedian(.x, w = w, na.rm = TRUE)
      },
      .complete = FALSE
    )
  ) %>% ungroup()

# Step 2. Denormalize the trend
df <- df %>%
  mutate(
    conc_denorm_RSV_WRM = pmax(conc_norm_RSV_WRM / flow, lod),
    
  )

results <- data.frame(
  niu = numeric(),
  var_delta_c = numeric(), 
  abs_dist_to_one = numeric()
)

for (niu_unknown in seq(0, 100, by = 0.1)) {
  
  df_tmp <- df %>%
    mutate(
      cv_RSV = sqrt(
        niu_unknown^2 * 1 / ((conc_denorm_RSV_WRM / d) * p * n) +
          (1 + niu_unknown^2) / (2 * p * n)
      ),
      delta_c_RSV = (conc_RSV - conc_denorm_RSV_WRM) /
        (cv_RSV * conc_denorm_RSV_WRM)
      
    )
  
  results <- rbind(
    results,
    data.frame(
      niu = niu_unknown,
      var_delta_c = var(df_tmp$delta_c_RSV, na.rm = TRUE),
      abs_dist_to_one = abs(1 - var(df_tmp$delta_c_RSV, na.rm = TRUE))
      
    )
  )
}

niu <- results[results$abs_dist_to_one == min(results$abs_dist_to_one),"niu"]

print(niu)
```

```r
# ──────────────────────────────────────────────────────────────────────────
# # Method to identify the outliers
# ──────────────────────────────────────────────────────────────────────────
identify_outliers <- function(df){
  df <- df %>% mutate(is_outlier_RSV = FALSE)
  df_return <- df %>% mutate(is_outlier_RSV = FALSE)

outlier_df <- data.frame()
    # remove outliers in iterations
    for (iter in 1:10){
  # Remove already identified outliers from the dataset
  # Step 1. Fit trend line
  # Calculate normalized concentrations conc_norm = conc * flow
  
  if (iter > 1) {outlier_df <- rbind(outlier_df, df[df$is_outlier_RSV == TRUE, ])}
  else if (iter == 1) {outlier_df <- df[df$is_outlier_RSV == TRUE, ]}
        
  df <- df[df$is_outlier_RSV == FALSE, ] %>%
    mutate(
      conc_norm_RSV = conc_RSV * flow,

    )
  df <- df %>%
    arrange(location, Date) %>%
    group_by(location) %>%
    mutate(
      conc_norm_RSV_WRM = slide_index2_dbl(
        .x = conc_norm_RSV, .y = Date, .i = Date,
        .before = days(window_size),
        .after  = days(window_size),
        .f = ~ {
          center <- .y[length(.y)]
          
          dist_days <- abs(as.numeric(.y - center))
          w <- max(dist_days) - dist_days + 1
          w <- w / sum(w)
          # weighted median
          weightedMedian(.x, w = w, na.rm = TRUE)
        },
        .complete = FALSE
      )
    ) %>% ungroup()




  # Step 2. Denormalize the trend
  df <- df %>%
    mutate(
      conc_denorm_RSV_WRM = pmax(conc_norm_RSV_WRM / flow, lod),

    )
  

  # Step 3. Predict the coefficient of variation for non-outlier dPCR measurements on each date
  df <- df  %>%

    mutate(

      cv_RSV = sqrt(niu^2 * 1/ ((conc_denorm_RSV_WRM / d) * p * n ) + (1 + niu^2)/(2 * p * n)) 
      
    )

  # Step 4. Standardise deviations
  df <- df %>%
    mutate(
      delta_c_RSV = (conc_RSV - conc_denorm_RSV_WRM) / (cv_RSV * conc_denorm_RSV_WRM)

    )

  # Step 5. Apply z-score threshold
  df <- df %>%
    mutate(
      is_outlier_RSV = ifelse(delta_c_RSV > z_threshold, 1, 0),

    )
  print(sum(df$is_outlier_RSV, na.rm = TRUE))
  if (sum(df$is_outlier_RSV, na.rm = TRUE) == 0) break

    }

  return (outlier_df)

}
```
