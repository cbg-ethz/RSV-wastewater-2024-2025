```r
# Helper function to build correlation annotation and plot per subtype
plot_entropy_vs_load <- function(df, subtype_label, title_label) {
  
  # Calculate Spearman correlations by genome region
  cor <- df %>%
    group_by(genome_region) %>%
    summarise(
      spearman_rho = round(cor.test(subtype_viral_load_raw, entropy, method = 'spearman')$estimate, 3),
      p_value = cor.test(subtype_viral_load_raw, entropy, method = 'spearman')$p.value
    )
  
  # Create text labels for correlation values
  cor_values <- data.frame(
    genome_region = cor$genome_region,
    rho = cor$spearman_rho,
    p_value = cor$p_value,
    x = rep(max(df$subtype_viral_load_raw, na.rm = TRUE) * 0.7, nrow(cor)),

    y = c(0.03, 0.028, 0.026) # 
            
  ) %>%

      mutate(label = paste0(genome_region, ":  rho = ", rho, ", p = ", signif(p_value, 2)))

  # Build ggplot
  ggplot(df, aes(x = subtype_viral_load_raw, y = entropy, color = genome_region)) +
    geom_point(alpha = 0.7, size = 3) +
    geom_smooth(method = "lm", se = FALSE, linewidth = 1) +
    geom_text(
      data = cor_values,
      aes(x = x, y = y, label = label, color = genome_region),
      inherit.aes = FALSE,
      size = 5,
      fontface = "bold"
    ) +
    scale_color_manual(values = c("F" = "#1f78b4", "G" = "#33a02c", "genome" = "#e31a1c")) +
    theme_bw() +
    labs(
      title = title_label,
      x = "Load (gc/person/day)",
      y = "Mean Shannon entropy",
      color = "Genome region"
    ) +
    theme(
      text = element_text(size = 18),
      axis.title = element_text(size = 20, face = "bold"),
      axis.text = element_text(size = 16),
      legend.title = element_text(size = 16, face = "bold"),
      legend.text = element_text(size = 14),
      plot.title = element_text(size = 20, face = "bold", hjust = 0.5),
      panel.grid.major = element_line(color = "grey90"),
      panel.grid.minor = element_blank()
    )
}
```

```r

model_with_pi_A <- lme(
  fixed = log_viral_loads ~ log_mean_whole_genome_pi + log_mean_positional_whole_genome_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location), #TODO adjust for irregular sampling
  data = diversity_data_sites_A,
  method = "ML"

)
model_without_pi_A <- lme(
  fixed = log_viral_loads ~ log_mean_positional_whole_genome_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location), #TODO adjust for irregular sampling
  data = diversity_data_sites_A,
  method = "ML"

)
model_with_pi_B <- lme(
  fixed = log_viral_loads ~ log_mean_whole_genome_pi + log_mean_positional_whole_genome_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location), #TODO adjust for irregular sampling
  data = diversity_data_sites_B,
  method = "ML"

)
model_without_pi_B <- lme(
  fixed = log_viral_loads ~  log_mean_positional_whole_genome_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location), #TODO adjust for irregular sampling
  data = diversity_data_sites_B,
  method = "ML"

)

summary(model_with_pi_A)
print("######################################################")
summary(model_without_pi_A)
print("######################################################")
anova(model_with_pi_A, model_without_pi_A)
print("######################################################")
summary(model_with_pi_B)
print("######################################################")
summary(model_without_pi_B)
print("######################################################")
anova(model_with_pi_B, model_without_pi_B)
```

# Viral loads association with piN and piS of the genome

```r
head(diversity_data)

diversity_data_piS_piN_modeling <- diversity_data %>%
  select(piS_piN, pi,viral_loads, fraction_covered_whole_genome, log_mean_whole_genome_codon_coverage) %>%

pivot_wider(
    names_from = piS_piN,
    values_from = pi
)
```

```r
head(diversity_data_piS_piN_modeling)
diversity_data_piS_piN_modeling$log_piS = log10(diversity_data_piS_piN_modeling$piS)
diversity_data_piS_piN_modeling$log_piN = log10(diversity_data_piS_piN_modeling$piN)

diversity_data_piS_piN_modeling <- diversity_data_piS_piN_modeling %>% drop_na()

diversity_data_piS_piN_modeling$log_viral_loads <- log10(diversity_data_piS_piN_modeling$viral_loads)
```

```r
diversity_data_piS_piN_modeling_A <- diversity_data_piS_piN_modeling[diversity_data_piS_piN_modeling$subtype == 'RSV_A', ]
diversity_data_piS_piN_modeling_B <- diversity_data_piS_piN_modeling[diversity_data_piS_piN_modeling$subtype == 'RSV_B', ]
```

```r
cor.test(diversity_data_piS_piN_modeling_A$log_piN, diversity_data_piS_piN_modeling_A$log_piS, method='spearman')
```

```r
cor.test(diversity_data_piS_piN_modeling$log_piN, diversity_data_piS_piN_modeling$log_piS, method='spearman')
cor.test(diversity_data_piS_piN_modeling$log_piN, diversity_data_piS_piN_modeling$log_piS, method='spearman')
```

```r
diversity_data_piS_piN_modeling_A <- diversity_data_piS_piN_modeling %>% filter(subtype == 'RSV_A')
diversity_data_piS_piN_modeling_B <- diversity_data_piS_piN_modeling %>% filter(subtype == 'RSV_B')
```

```r

model_A_with_piS <- lme(
  fixed = log_viral_loads ~ log_piS + fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location),
  data = diversity_data_piS_piN_modeling_A,
  method = "ML")

model_A_with_piN <- lme(
  fixed = log_viral_loads ~ log_piN + fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location),
  data = diversity_data_piS_piN_modeling_A,
  method = "ML")


model_A_without_pi <- lme(
  fixed = log_viral_loads ~ + fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location),
  data = diversity_data_piS_piN_modeling_A,
  method = "ML")

summary(model_A_with_piS)
print("######################################################")
summary(model_A_with_piN)
print("######################################################")
summary(model_A_without_pi)
print("######################################################")
```

```r
AIC_with  <- AIC(model_A_with_piS)
AIC_without <- AIC(model_A_without_pi)

delta_AIC <- AIC_without - AIC_with
delta_AIC
BIC_with  <- BIC(model_A_with_piS)
BIC_without <- BIC(model_A_without_pi)

delta_BIC <- BIC_without - BIC_with
delta_BIC
```

```r
AIC_with  <- AIC(model_A_with_piN)
AIC_without <- AIC(model_A_without_pi)

delta_AIC <- AIC_without - AIC_with
delta_AIC
BIC_with  <- BIC(model_A_with_piN)
BIC_without <- BIC(model_A_without_pi)

delta_BIC <- BIC_without - BIC_with
delta_BIC
```

```r
model_B_with_piS <- lme(
  fixed = log_viral_loads ~ log_piS + fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location), 
  data = diversity_data_piS_piN_modeling_B,
  method = "ML")


model_B_with_piN <- lme(
  fixed = log_viral_loads ~ log_piN + fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location), 
  data = diversity_data_piS_piN_modeling_B,
  method = "ML")



model_B_without_pi <- lme(
  fixed = log_viral_loads ~ fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location),
  data = diversity_data_piS_piN_modeling_B,
  method = "ML")

summary(model_B_with_piN)
print("######################################################")
summary(model_B_with_piS)
print("######################################################")
summary(model_B_without_pi)

anova(model_B_with_piN,model_B_without_pi)
anova(model_B_with_piS,model_B_without_pi)
```

```r
AIC_with  <- AIC(model_B_with_piS)
AIC_without <- AIC(model_B_without_pi)

delta_AIC <- AIC_without - AIC_with
delta_AIC
BIC_with  <- BIC(model_B_with_piS)
BIC_without <- BIC(model_B_without_pi)

delta_BIC <- BIC_without - BIC_with
delta_BIC
```

```r
AIC_with  <- AIC(model_B_with_piN)
AIC_without <- AIC(model_B_without_pi)

delta_AIC <- AIC_without - AIC_with
delta_AIC
BIC_with  <- BIC(model_B_with_piN)
BIC_without <- BIC(model_B_without_pi)

delta_BIC <- BIC_without - BIC_with
delta_BIC
```
