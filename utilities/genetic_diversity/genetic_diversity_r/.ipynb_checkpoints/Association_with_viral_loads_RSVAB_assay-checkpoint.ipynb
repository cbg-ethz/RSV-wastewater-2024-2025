```r
codon_coverage_merged_RSVA <- read_csv( "../../../RSV/data_analysis/data/RSVA_2024_2025/selective_pressures/results_snpgenie_processed/RSVA_codon_coverage_merged.csv",
                                      show_col_types = FALSE)
codon_coverage_merged_RSVB <- read_csv( "../../../RSV/data_analysis/data/RSVB_2024_2025/selective_pressures/results_snpgenie_processed/RSVB_codon_coverage_merged.csv",
                                      show_col_types = FALSE)

RSVA_site_nuc_diversity_whole_genome <- read_csv("../../../RSV/data_analysis/data/RSVA_2024_2025/selective_pressures/results_snpgenie_processed/RSVA_site_nuc_diversity_whole_genome.csv",
                                   show_col_types = FALSE)
RSVB_site_nuc_diversity_whole_genome <- read_csv("../../../RSV/data_analysis/data/RSVB_2024_2025/selective_pressures/results_snpgenie_processed/RSVB_site_nuc_diversity_whole_genome.csv",
                                   show_col_types = FALSE)
```

```r
# Site coverage
site_nuc_diversity_merged_whole_genome <- rbind(RSVA_site_nuc_diversity_whole_genome, RSVB_site_nuc_diversity_whole_genome)
head(site_nuc_diversity_merged_whole_genome)
site_nuc_diversity_merged_whole_genome[(site_nuc_diversity_merged_whole_genome$reference =="EPI_ISL_412866"), "reference"] <- "RSV_A"
site_nuc_diversity_merged_whole_genome[(site_nuc_diversity_merged_whole_genome$reference =="EPI_ISL_1653999"), "reference"] <- "RSV_B"

site_nuc_diversity_merged_whole_genome <- site_nuc_diversity_merged_whole_genome %>%
mutate(log_mean_whole_genome_pi = log10(mean_whole_genome_pi))

site_nuc_diversity_merged_whole_genome['location'] <- sapply(strsplit(site_nuc_diversity_merged_whole_genome$sample, "_"), `[`, 2)


head(site_nuc_diversity_merged_whole_genome)
```

```r
# Codon coverage

codon_coverage_merged <- rbind(codon_coverage_merged_RSVA, codon_coverage_merged_RSVB)
# replace reference names by subtype names
codon_coverage_merged[(codon_coverage_merged$reference =="EPI_ISL_412866"), "reference"] <- "RSV_A"
codon_coverage_merged[(codon_coverage_merged$reference =="EPI_ISL_1653999"), "reference"] <- "RSV_B"

dim(codon_coverage_merged)
head(codon_coverage_merged)
```

```r
# Taking only samples where genome is covered at least 50%
codon_coverage_merged <- codon_coverage_merged %>% filter(fraction_covered_whole_genome >= FRACTION_COVERED_THRESHOLD)
codon_coverage_merged <- codon_coverage_merged %>% filter(codon_coverage >= COVERAGE_THRESHOLD)

codon_coverage_merged <- codon_coverage_merged %>% group_by(sample, reference) %>%
mutate(mean_whole_genome_codon_coverage = mean(codon_coverage))

dim(codon_coverage_merged)

```

# Viral loads

```r
# Read the concentration data
samples <- read.csv("/Users/arimaite/Documents/GitHub/combined_virus_sequencing/RSV/data_analysis/data/samples_tsv/samples_corrected_v2.tsv", 
                    sep = '\t',
                    header = TRUE,
                    colClasses = c(
                      "location" = "character"
                    ))


samples$location <- gsub("10", "Zurich", samples$location)
samples$location <- gsub("16", "Geneva", samples$location)
samples$location <- gsub("15", "Basel", samples$location)
samples$location <- gsub("05", "Lugano", samples$location)
samples$location <- gsub("25", "Laupen", samples$location)
samples$location <- gsub("17", "Chur", samples$location)
samples$date <- as.character(samples$date)

# Have a list of samples that were sequenced for RSV-A (remove only B), and remove non-specific binding:
rsvb_samples <- samples %>% 
  filter(!grepl("A", samples$protocol))
rsvb_samples$date <- as.character(rsvb_samples$date)  # Convert to Date format

# Have a list of samples that were sequenced for RSV-A (remove only B), and remove non-specific binding:
rsva_samples <- samples %>% 
  filter(!grepl("B", samples$protocol))
rsva_samples$date <- as.character(rsva_samples$date)  # Convert to Date format
```

```r
# Get the concentrations from RSV-A/B assay
rsv_ab_assay_raw <- read.csv(file = '/Users/arimaite/Documents/GitHub/combined_virus_sequencing/RSV/data_analysis/data/concentrations/rsv_ab_assay/viral_loads_imputated_rsvab.csv',
                         check.names = FALSE) %>%
  select(-c(1)) %>%
  select(Date, location, RSV_A, RSV_B) %>% # Using loads - these are concentrations normalized by the flow and population size
  mutate(Date = as.Date(Date, "%Y-%m-%d")) %>%
  rename(date = Date) %>% 
  pivot_longer(
    c(RSV_A, RSV_B),
    names_to = "subtype",
    values_to = "subtype_viral_load_raw"
  ) 

```

```r
head(codon_coverage_merged)
head(rsv_ab_assay_raw)
```

```r
lookup = c(
    '05'='Lugano',
    '10'='Zurich',
    '15'='Basel',
    '16'='Geneva',
    '17'='Chur',
    '25'='Laupen'
)
codon_coverage_merged$location <- lookup[codon_coverage_merged$location]
codon_coverage_merged <- codon_coverage_merged %>% rename(subtype = reference)
head(codon_coverage_merged)
head(rsv_ab_assay_raw)
```

```r
diversity_data <- inner_join(rsv_ab_assay_raw, codon_coverage_merged, by = c("date", "location", "subtype"))
head(diversity_data)
dim(diversity_data)
```

```r
diversity_data <- diversity_data %>%
  group_by(subtype, location, date) %>%
    mutate(
        piS = sum(S_diffs, na.rm = TRUE) / sum(S_sites, na.rm = TRUE),
        piN = sum(N_diffs, na.rm = TRUE) / sum(N_sites, na.rm = TRUE),
        ratio = piN - piS,
        mean_fraction_covered_whole_genome = mean(fraction_covered_whole_genome)
    ) %>%
    distinct(sample,
             location,
             subtype,
            # product,
             mean_fraction_covered_whole_genome,
             mean_whole_genome_codon_coverage, 
             piS,
             piN, 
             ratio,
             date,
             subtype_viral_load_raw
            ) %>%
      mutate(location = factor(location)) %>%
      mutate(date = as.Date(date, format = "%Y_%m_%d")) %>%
mutate(log_mean_whole_genome_codon_coverage = log10(mean_whole_genome_codon_coverage))

head(diversity_data)
dim(diversity_data)
```

```r

diversity_data <- diversity_data %>%
#group_by(
#    subtype, location, date, subtype_viral_load_raw
#) %>% 
    pivot_longer(cols = c("piS", "piN"), names_to = "piS_piN", values_to = "pi")

head(diversity_data)
```

```r
head(diversity_data)
diversity_data <- diversity_data %>% pivot_wider(
    names_from = piS_piN,
    values_from = pi
)
```

```r
head(diversity_data)
diversity_data$log_piS = log10(diversity_data$piS)
diversity_data$log_piN = log10(diversity_data$piN)

diversity_data<- diversity_data %>% drop_na()

diversity_data$log_subtype_viral_load_raw <- log10(diversity_data$subtype_viral_load_raw)
```

```r
cor.test(diversity_data$log_piN, diversity_data$log_piS)
```

```r
model_with_piN <- lme(
  fixed = log_subtype_viral_load_raw ~ log_piN * subtype + mean_fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  #correlation = corCAR1(form = ~ as.numeric(date) | location/subtype), #TODO adjust for irregular sampling
  data = diversity_data,
  method = "ML")

model_with_piS <- lme(
  fixed = log_subtype_viral_load_raw ~ log_piS * subtype + mean_fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  #correlation = corCAR1(form = ~ as.numeric(date) | location/subtype), #TODO adjust for irregular sampling
  data = diversity_data,
  method = "ML")


model_without_pi <- lme(
  fixed = log_subtype_viral_load_raw ~ subtype + mean_fraction_covered_whole_genome + log_mean_whole_genome_codon_coverage,
  random = ~ 1 | location,
  #correlation = corCAR1(form = ~ as.numeric(date) | location/subtype), #TODO adjust for irregular sampling
  data = diversity_data,
  method = "ML")

summary(model_with_piN)

summary(model_with_piS)

summary(model_without_pi)
```
