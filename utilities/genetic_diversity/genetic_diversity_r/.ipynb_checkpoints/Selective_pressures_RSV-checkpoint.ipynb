# Compare selective pressures between different genes

```r
# To estimate selective pressures of different genes, we estimated whether piN-piS difference significantly differs from zero
# in each of the genes.
# Only using the samples with at least 50% of the genome covered by more than 10x reads.
# To get confidence intervals we implement hierarchical  (two-level) bootstrap: first, resample samples with replacement,
# then resample codons with replacement.
# To simplify the procedure (and reduce time complexity), we assume independence between samples. However, we have time-series spatially clustered samples, therefore
# estimating block bootstrap could be useful.


# The procedure is as follows:
# 
# 1. Running SNPGenie and extracting the results for each codon.
# We resampled 100 times to get the sets of the samples.
# 
# 2. For each gene do 1000 times:
#        Resample # of gene length (in codons) codons with replacement. Only covered codons are used.
#        Estimate piN and piS parameters and their difference theta = piN - piS
#        Store this values in a list
# 3. Estimate standard deviation across replicates SE(theta)

```

# Estimating selective pressures

```r
estimate_selective_pressures <- function(product,
                                         ref_genome,
                                         start_position_reference_product,
                                         end_position_reference_product,
                                         window_size = 1){
    reference_product_all_windows <- list()
    reference_product_all_samples <- list()
    
# window_size in codons
    window_size <- window_size * 3 # window size in nucleotide bases
    n <- (end_position_reference_product - start_position_reference_product) %/% window_size

samples <- codon_coverage_merged %>%
  filter(product == product, reference == ref_genome) %>%
  distinct(sample) %>% 
  pull(sample)
    
for (j in seq_along(samples)) {
  s <- samples[j]
    
    codon_coverage_merged_reference_product <- codon_coverage_merged %>% filter(product == product, 
                                       reference == ref_genome, sample == s)
    for (i in seq(0, n - 1)) {
    start <- start_position_reference_product + i * window_size
    end <- start + window_size
    selected_window <- codon_coverage_merged_reference_product[
        ((codon_coverage_merged_reference_product["site1"] >= start) & (codon_coverage_merged_reference_product["site1"] < end)), ]
        
    selected_window <- selected_window %>%
    mutate(
        piS = ifelse (sum(S_sites, na.rm = TRUE) != 0, sum(S_diffs, na.rm = TRUE) / sum(S_sites, na.rm = TRUE), 0),
        piN = ifelse (sum(N_sites, na.rm = TRUE) != 0, sum(N_diffs, na.rm = TRUE) / sum(N_sites, na.rm = TRUE), 0),
        ratio = piN - piS,
        start_in_codon = (start - start_position_reference_product) / 3, # start position in codon coordinates
        end_in_codon = (end - start_position_reference_product) / 3
    ) %>%
    distinct(sample,
             location,
             product,
             reference,
             mean_positional_coverage, 
             piS,
             piN, 
             ratio,
             date,
             start_in_codon,
             end_in_codon,
             codon_coverage
            ) %>%
      mutate(location = factor(location)) %>%
      mutate(date = as.Date(date, format = "%Y_%m_%d"))
    reference_product_all_windows[[i+1]] <- selected_window
}
    reference_product_all_windows_df <- bind_rows(reference_product_all_windows)
    reference_product_all_samples[[j]] <- reference_product_all_windows_df
}

reference_product_all_samples_df <- bind_rows(reference_product_all_samples)

    return(reference_product_all_samples_df)
    
}
```

# Hotspots identification: fused lasso approach

```r
fit_fused_lasso <- function(data_frame){

# Fit mixed effects model to find the conditional mean of the diversity for each codon position.
    
    model <- lme(
      fixed = piN ~ 0 + factor(end_in_codon),
      random = ~ 1 | location,
      correlation = corCAR1(form = ~ as.numeric(date) | location/end_in_codon), 
      data = data_frame,
      method = "ML"
    
    )
    # choose the beta's coefficients of the fixed effects of the positions
    beta_fixed_effects <- fixef(model)
    beta_i <- beta_fixed_effects[grep("factor\\(end_in_codon\\)", names(beta_fixed_effects))]
    positions <- as.numeric(gsub("factor\\(end_in_codon\\)", "", names(beta_i)))
    beta_vec <- beta_i[order(positions)]
    
    beta_vec[beta_vec < 0] <- 0



n <- length(beta_vec)
D <- diff(diag(n))

fit <- fusedlasso(beta_vec, D = D)
    return(list(fit = fit,
               beta_vec = beta_vec))

    
}


identify_elevated_diversity_regions <- function(lasso_model,
                                                beta_vec, 
                                                data_frame_2, 
                                                target_lambda,
                                                gene
                                               ) {
    

print(colnames(lasso_model$beta))
which_lambda <- which(colnames(lasso_model$beta) == target_lambda)[1]
# which(fit$beta =lambda,) take the first
print(colnames(lasso_model$beta)[which_lambda])



# Extract the results from the fused lasso model
    
beta_hat <- lasso_model$beta[, which_lambda]

pos <- data_frame_2$midpoint   # or codon_position

# Plot
    if (gene == 'F') {
        threshold = quantile(beta_hat, 0.90)
    } else if (gene == 'G') {
        threshold = quantile(beta_hat, 0.50)
    } else {
         warning("Unknown gene provided. Threshold not defined.")
         threshold <- NA
    }

    
plot_df <- data.frame(
  pos = pos,
  beta_raw = beta_vec,
  beta_smooth = beta_hat,
  high = beta_hat >= threshold
)
plot_df <- plot_df %>%
  mutate(region = cumsum(high != lag(high, default = first(high))))
 
    
  regions <- plot_df %>%
  filter(high) %>%
  group_by(region) %>%
  summarise(start = min(pos),
            end   = max(pos),
            .groups = "drop")

plot <- ggplot(plot_df, aes(x = pos)) +

  # Blue shaded regions
  geom_rect(data = regions,
            aes(xmin = start,
                xmax = end,
                ymin = -Inf,
                ymax = Inf),
            fill = "#4A90E2",
            alpha = 0.15,
            inherit.aes = FALSE) +

  # Raw estimates (small grey dots)
  geom_point(aes(y = beta_raw),
             size = 1,
             color = "grey60") +

  # Smoothed line
  geom_line(aes(y = beta_smooth),
            size = 1,
            color = "#D62728") +

  # Threshold line
  geom_hline(yintercept = threshold,
             linetype = "dashed",
             color = "black") +

  labs(
    x = "Codon Position",
    y = "piN",
    title = "",
  ) +
  scale_x_continuous(
    breaks = seq(
      floor(min(plot_df$pos, na.rm = TRUE)),
      ceiling(max(plot_df$pos, na.rm = TRUE)),
      by = 50
    )
  ) +
     theme_bw(base_size = 20) +
  theme(
    plot.title = element_text(face = "bold"),
    axis.line = element_line(color = "black")
  )

    print(plot_df)
    return(list(
        regions = regions, 
        plot = plot))
}
```

# Estimating selective pressures

```r
model_piN <- lme(
  fixed = piN ~ 0 + factor(end_in_codon),
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location/end_in_codon), 
  data = data_frame,
  method = "ML"
)
# choose the beta's coefficients of the fixed effects of the positions
beta_fixed_effects_piN <- fixef(model_piN)
beta_i_piN <- beta_fixed_effects_piN[grep("factor\\(end_in_codon\\)", names(beta_fixed_effects_piN))]
positions_piN <- as.numeric(gsub("factor\\(end_in_codon\\)", "", names(beta_i_piN)))
beta_vec_piN <- beta_i_piN[order(positions_piN)]

beta_vec_piN[beta_vec_piN<0] <- 0
```

```r
model_piS <- lme(
  fixed = piS ~ 0 + factor(end_in_codon),
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location/end_in_codon), 
  data = data_frame,
  method = "ML"
)
# choose the beta's coefficients of the fixed effects of the positions
beta_fixed_effects_piS <- fixef(model_piS)
beta_i_piS <- beta_fixed_effects_piS[grep("factor\\(end_in_codon\\)", names(beta_fixed_effects_piS))]
positions_piS <- as.numeric(gsub("factor\\(end_in_codon\\)", "", names(beta_i_piS)))
beta_vec_piS <- beta_i_piS[order(positions_piS)]

beta_vec_piS[beta_vec_piS<0] <- 0
```

# Estimating the difference piN - piS. 

```r
beta_vec <- beta_vec_piN - beta_vec_piS
n <- length(beta_vec)
D <- diff(diag(n))
fit_difference <- fusedlasso(beta_vec, D = D)
```

# RSV-A G gene

```r
RSV_A_G_all_samples_df <- estimate_selective_pressures("G",
                             "RSV_A",
                             4652,
                             5617,
                             1)
```

```r
RSVA_G_windows_df <- RSV_A_G_all_samples_df %>%
mutate(
    midpoint = (start_in_codon + end_in_codon) / 2) 
RSVA_G_windows_df <- RSVA_G_windows_df %>% 
group_by(midpoint) %>% 
summarize(m_piS = mean(piS),
          m_piN = mean(piN),
          m_ratio = mean(ratio),
          .groups = "drop" )
RSVA_G_long <- RSVA_G_windows_df %>% select(midpoint, 
                                            m_piS,
                                            m_piN, 
                                            m_ratio) %>%
#mutate(diff = piN - piS)
pivot_longer(cols = c(m_piS, m_piN),
             names_to = "type",
             values_to = "pi")
```

```r
options(repr.plot.width = 15, repr.plot.height =4)



ratio_scale <- max(RSVA_G_windows_df$m_piN, na.rm = TRUE) /
               max(abs(RSVA_G_windows_df$m_ratio), na.rm = TRUE)
ggplot() +
  # πS and πN
  geom_line(data = RSVA_G_long,
            aes(x = midpoint, y = pi, color = type),
            linewidth = 1) +
  geom_point(data = RSVA_G_long,
             aes(x = midpoint, y = pi, color = type),
             alpha = 0.6) +
  
  # ratio (scaled)
  geom_line(data = RSVA_G_windows_df,
            aes(x = midpoint,
                y = m_ratio * ratio_scale),
            color = "black",
            linewidth = 1#,
            #linetype = "dashed"
           ) +

    geom_smooth(
    data = RSVA_G_long,
    aes(x = midpoint, y = pi, color = type),
    method = "loess",
  span = 0.3,
    se = FALSE,
    linewidth = 1#,
        #linetype = "dashed"
  ) +

  scale_y_continuous(
    name = "Mean nucleotide diversity",
    sec.axis = sec_axis(~ . / ratio_scale,
                        name = "piN-piS")
  ) +
  
  scale_x_continuous(
    breaks = seq(
      floor(min(RSVA_G_long$midpoint, na.rm = TRUE)),
      ceiling(max(RSVA_G_long$midpoint, na.rm = TRUE)),
      by = 50
    )
  ) +
  
  scale_color_manual(values = c("m_piS" = "blue",
                                "m_piN" = "red"),
                     labels = c("piS"),
                                "piN") +
  
  theme_bw() +
  labs(
    title = "Sliding window nucleotide diversity across RSV-A G gene",
    x = "Genome position",
    color = ""
  ) +
  theme(text = element_text(size = 15))
```

```r
lasso_fit_RSVA_G_model <- fit_fused_lasso(RSV_A_G_all_samples_df)
```

```r
RSV_A_G_results <- identify_elevated_diversity_regions(
    beta_vec = lasso_fit_RSVA_G_model$beta_vec, 
    lasso_model = lasso_fit_RSVA_G_model$fit,
    data_frame_2 = RSVA_G_windows_df, 
    target_lambda = 0.017,
    gene = 'G')
```

```r
options(repr.plot.width = 10, repr.plot.height =5)

print(RSV_A_G_results$plot)
```

# RSV-B F gene

```r
RSV_B_F_all_samples_df <- estimate_selective_pressures("F",
                             "RSV_B",
                             5676,
                             7400,
                             1)
```

```r
RSVB_F_windows_df <- RSV_B_F_all_samples_df %>%
mutate(
    midpoint = (start_in_codon + end_in_codon) / 2) 
RSVB_F_windows_df <- RSVB_F_windows_df %>% 
group_by(midpoint) %>% 
summarize(m_piS = mean(piS),
          m_piN = mean(piN),
          m_ratio = mean(ratio),
          .groups = "drop" )
RSVB_F_long <- RSVB_F_windows_df %>% select(midpoint, 
                                            m_piS,
                                            m_piN, 
                                            m_ratio) %>%
#mutate(diff = piN - piS)
pivot_longer(cols = c(m_piS, m_piN),
             names_to = "type",
             values_to = "pi")
```

```r
options(repr.plot.width = 15, repr.plot.height =4)



ratio_scale <- max(RSVB_F_windows_df$m_piN, na.rm = TRUE) /
               max(abs(RSVB_F_windows_df$m_ratio), na.rm = TRUE)
ggplot() +
  # πS and πN
  geom_line(data = RSVB_F_long,
            aes(x = midpoint, y = pi, color = type),
            linewidth = 1) +
  geom_point(data = RSVB_F_long,
             aes(x = midpoint, y = pi, color = type),
             alpha = 0.6) +
  
  # ratio (scaled)
  geom_line(data = RSVB_F_windows_df,
            aes(x = midpoint,
                y = m_ratio * ratio_scale),
            color = "black",
            linewidth = 1#,
            #linetype = "dashed"
           ) +

    geom_smooth(
    data = RSVB_F_long,
    aes(x = midpoint, y = pi, color = type),
    method = "loess",
  span = 0.3,
    se = FALSE,
    linewidth = 1
  ) +

  scale_y_continuous(
    name = "Mean nucleotide diversity",
    sec.axis = sec_axis(~ . / ratio_scale,
                        name = "piN-piS")
  ) +
  
  scale_x_continuous(
    breaks = seq(
      floor(min(RSVB_F_long$midpoint, na.rm = TRUE)),
      ceiling(max(RSVB_F_long$midpoint, na.rm = TRUE)),
      by = 50
    )
  ) +
  
  scale_color_manual(values = c("m_piS" = "blue",
                                "m_piN" = "red")) +
  
  theme_bw() +
  labs(
    title = "Sliding window nucleotide diversity across RSV-B F gene",
    x = "Genome position",
    color = ""
  ) +
  theme(text = element_text(size = 15))
```

```r
lasso_fit_RSVB_F_model <- fit_fused_lasso(RSV_B_F_all_samples_df)
```

```r
RSV_B_F_results <- identify_elevated_diversity_regions(
    beta_vec = lasso_fit_RSVB_F_model$beta_vec, 
    lasso_model = lasso_fit_RSVB_F_model$fit,
    data_frame_2 = RSVB_F_windows_df, 
    target_lambda = 0.008,
    gene = 'F')
```

```r
options(repr.plot.width = 10, repr.plot.height =5)

print(RSV_B_F_results$plot)
```
