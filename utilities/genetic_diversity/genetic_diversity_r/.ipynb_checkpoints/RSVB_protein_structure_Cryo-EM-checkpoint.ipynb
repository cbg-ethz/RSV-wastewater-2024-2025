```r
# Taking only samples where corresponding gene is covered at least 50%
codon_coverage_merged <- codon_coverage_merged %>% filter(fraction_covered >= 0.50)
codon_coverage_merged <- codon_coverage_merged %>% filter(fraction_covered_whole_genome >= 0.5)
codon_coverage_merged <- codon_coverage_merged %>% filter(codon_coverage >= 100)

dim(codon_coverage_merged)

```

```r
estimate_selective_pressures <- function(product,
                                         ref_genome,
                                         start_position_reference_product,
                                         end_position_reference_product,
                                         window_size = 1){
    reference_product_all_windows <- list()
    reference_product_all_samples <- list()
    
# window_size in codons
    window_size <- window_size * 3 # window size in nucleotide bases
    n <- (end_position_reference_product - start_position_reference_product) %/% window_size

samples <- codon_coverage_merged %>%
  filter(product == product, reference == ref_genome) %>%
  distinct(sample) %>% 
  pull(sample)
    
for (j in seq_along(samples)) {
  s <- samples[j]
    
    codon_coverage_merged_reference_product <- codon_coverage_merged %>% filter(product == product, 
                                       reference == ref_genome, sample == s)
    for (i in seq(0, n - 1)) {
    start <- start_position_reference_product + i * window_size
    end <- start + window_size
    selected_window <- codon_coverage_merged_reference_product[
        ((codon_coverage_merged_reference_product["site1"] >= start) & (codon_coverage_merged_reference_product["site1"] < end)), ]
        
    selected_window <- selected_window %>%
    mutate(
        piS = ifelse (sum(S_sites, na.rm = TRUE) != 0, sum(S_diffs, na.rm = TRUE) / sum(S_sites, na.rm = TRUE), 0),
        piN = ifelse (sum(N_sites, na.rm = TRUE) != 0, sum(N_diffs, na.rm = TRUE) / sum(N_sites, na.rm = TRUE), 0),
        ratio = piN - piS,
        start_in_codon = (start - start_position_reference_product) / 3, # start position in codon coordinates
        end_in_codon = (end - start_position_reference_product) / 3
    ) %>%
    distinct(sample,
             location,
             product,
             reference,
             mean_positional_coverage, 
             piS,
             piN, 
             ratio,
             date,
             start_in_codon,
             end_in_codon,
             codon_coverage
            ) %>%
      mutate(location = factor(location)) %>%
      mutate(date = as.Date(date, format = "%Y_%m_%d"))
      reference_product_all_windows[[i+1]] <- selected_window
    }
    
    reference_product_all_windows_df <- bind_rows(reference_product_all_windows)
    reference_product_all_samples[[j]] <- reference_product_all_windows_df
    }

    reference_product_all_samples_df <- bind_rows(reference_product_all_samples)

    return(reference_product_all_samples_df)
    
}
```

```r
RSV_B_F_all_samples_df <- estimate_selective_pressures("F",
                             "RSV_B",
                             5676,
                             7400,
                             1)

```

```r
model <- lme(
  fixed = ratio ~ 0 + factor(end_in_codon),
  random = ~ 1 | location,
  correlation = corCAR1(form = ~ as.numeric(date) | location/end_in_codon), 
  data = RSV_A_F_all_samples_df,
  method = "ML"

)
# choose the beta's coefficients of the fixed effects of the positions
beta_fixed_effects <- fixef(model)
beta_i <- beta_fixed_effects[grep("factor\\(end_in_codon\\)", names(beta_fixed_effects))]
positions <- as.numeric(gsub("factor\\(end_in_codon\\)", "", names(beta_i)))
beta_vec <- beta_i[sort(positions)]



```

```r

pdb <- read.pdb("../../../RSV/resources/PDB_protein_structure/pdb9b2x.ent")
```

```r
ca <- atom.select(pdb, chain = "A", elety = "CA")
```

```r
coords <- pdb$xyz[ca$xyz]
#coords
```

```r
# Convert to matrix
coords <- matrix(coords, ncol = 3, byrow = TRUE)
residues <- pdb$atom$resno[ca$atom]

rownames(coords) <- residues
colnames(coords) <- c("x", "y", "z")
head(coords)
coords <- as.data.frame(coords)
coords$rownames <- as.integer(rownames(coords))
```

```r
dist_mat <- as.matrix(dist(coords)) # distance matrix with the distances in angstroms
# If the distance is larger than 10 A, we assume that there is no similarity in selective pressures values between the amino acids.
# => set the distance to Inf
cutoff <- 30 # angstroms

#dist_mat[(dist_mat > cutoff)] = Inf 



# define Gaussian smoothing function
kernel_smoothing <- function(distance, bandwidth){
    
    return(exp(-(distance^2)/(2*bandwidth)))
    
}
b <- 100
# Define similarity matrix S
S <- kernel_smoothing(dist_mat, b)
S <- S / rowSums(S, na.rm = TRUE)
head(S)
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)

heatmap(
  S,
  Rowv = NA,
  Colv = NA,
  dendrogram = "none",
  scale = "none",
  col = colorRampPalette(c("white", "black"))(100),
  cexRow = 1.2,   # increase row label size
  cexCol = 1.2    # increase column label size
)
```

```r
#### Define degree matrix D
D <- diag(rowSums(S))
colnames(D) <- colnames(S)
rownames(D) <- rownames(S)
head(D)
```

```r
# Laplacian matrix L
L <- D - S
```

```r
merged <- right_join(plot_df, coords, by = c("pos" = "rownames"))
head(merged)
tail(merged)
```

```r
dim(L) 
dim(merged)
```

```r
lambda <- 1   # tune this
I <- diag(nrow(L))
beta_smooth <- solve(I + lambda * L, merged$beta_vec)
```

```r
options(repr.plot.width = 15, repr.plot.height =5)
library(ggplot2)

plot_df <- data.frame(
  position = merged$pos,
  beta_smooth = beta_smooth
)

ggplot(plot_df, aes(x = position, y = beta_smooth)) +
    geom_point(size = 1.2) +
    scale_x_continuous(
    breaks = seq(
      from = floor(min(plot_df$position) / 50) * 50,
      to   = ceiling(max(plot_df$position) / 50) * 50,
      by   = 50
    )
  ) +
  labs(
    title = "Smoothed piN-piS Across RSV-B Genome",
    x = "Genomic Position",
    y = "piN - piS "
  ) +
  theme_minimal(base_size = 18) +
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),
    axis.title = element_text(face = "bold"),
    axis.text.x = element_text(angle = 45, hjust = 1),
    panel.grid.minor = element_blank()
  )
```
