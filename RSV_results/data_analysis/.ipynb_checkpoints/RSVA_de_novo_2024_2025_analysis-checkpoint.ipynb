```r
# remove rows that are completely missing
mut_freq_ZH<- mut_freq_ZH[rowSums(mut_freq_ZH, na.rm = TRUE)!=0, ]
mut_freq_GE<- mut_freq_GE[rowSums(mut_freq_GE, na.rm = TRUE)!=0, ]
mut_freq_LUG<- mut_freq_LUG[rowSums(mut_freq_LUG, na.rm = TRUE)!=0, ]
mut_freq_CHUR<- mut_freq_CHUR[rowSums(mut_freq_CHUR, na.rm = TRUE)!=0, ]

bandwidths <- c(3, 5, 7, 10, 15)

for (bw in bandwidths) {
    
  # Smooth data
  mut_freq_ZH_smoothed <- smooth_data_matrix(df = mut_freq_ZH, bandwidth = bw)
  mut_freq_GE_smoothed <- smooth_data_matrix(df = mut_freq_GE, bandwidth = bw)
  mut_freq_LUG_smoothed <- smooth_data_matrix(df = mut_freq_LUG, bandwidth = bw)
  mut_freq_CHUR_smoothed <- smooth_data_matrix(df = mut_freq_CHUR, bandwidth = bw)


  # Save individual files
  write.csv(mut_freq_ZH_smoothed, paste0("data/RSVA_2024_2025/smoothed_mut_frequencies/RSVA_ZH_", bw, ".csv"), row.names = TRUE)
  write.csv(mut_freq_GE_smoothed, paste0("data/RSVA_2024_2025/smoothed_mut_frequencies/RSVA_GE_", bw, ".csv"), row.names = TRUE)
  write.csv(mut_freq_LUG_smoothed, paste0("data/RSVA_2024_2025/smoothed_mut_frequencies/RSVA_LUGANO_", bw, ".csv"), row.names = TRUE)
  write.csv(mut_freq_CHUR_smoothed, paste0("data/RSVA_2024_2025/smoothed_mut_frequencies/RSVA_CHUR_", bw, ".csv"), row.names = TRUE)

  # Merge data: Zurich and Geneva
  combined_mut_freq_ZH_GE <- merge(mut_freq_ZH_smoothed, mut_freq_GE_smoothed, by = "row.names", all = FALSE)

  # Restore row names and remove extra column
  rownames(combined_mut_freq_ZH_GE) <- combined_mut_freq_ZH_GE$Row.names
  combined_mut_freq_ZH_GE$Row.names <- NULL

  # Save merged file
  write.csv(combined_mut_freq_ZH_GE, paste0("data/RSVA_2024_2025/smoothed_mut_frequencies/RSVA_ZH_GE_", bw, ".csv"), row.names = TRUE)




  # Merge data: Zurich, Geneva, Lugano and Chur

  combined_mut_freq_ZH_GE_LUG_CHUR <- merge(combined_mut_freq_ZH_GE, mut_freq_LUG_smoothed, by = "row.names", all = FALSE)
     # Restore row names and remove extra column
  rownames(combined_mut_freq_ZH_GE_LUG_CHUR) <- combined_mut_freq_ZH_GE_LUG_CHUR$Row.names
  combined_mut_freq_ZH_GE_LUG_CHUR$Row.names <- NULL
    
  combined_mut_freq_ZH_GE_LUG_CHUR <- merge(combined_mut_freq_ZH_GE_LUG_CHUR, mut_freq_CHUR_smoothed, by = "row.names", all = FALSE)

  # Restore row names and remove extra column
  rownames(combined_mut_freq_ZH_GE_LUG_CHUR) <- combined_mut_freq_ZH_GE_LUG_CHUR$Row.names
  combined_mut_freq_ZH_GE_LUG_CHUR$Row.names <- NULL

  # Save merged file
  write.csv(combined_mut_freq_ZH_GE_LUG_CHUR, paste0("data/RSVA_2024_2025/smoothed_mut_frequencies/RSVA_ZH_GE_LUG_CHUR_", bw, ".csv"), row.names = TRUE)
}
```

```r
# Define the base path
base_path <- "data/RSVA_2024_2025/cross_validation/"

# List of all CSV filenames
csv_files <- c(
  "output_RSVA_ZH_5.csv", "output_RSVA_CHUR_3.csv", "output_RSVA_CHUR_5.csv",
  "output_RSVA_CHUR_7.csv", "output_RSVA_CHUR_10.csv", "output_RSVA_CHUR_15.csv",
  "output_RSVA_GE_3.csv", "output_RSVA_GE_5.csv", "output_RSVA_GE_7.csv",
  "output_RSVA_GE_10.csv", "output_RSVA_GE_15.csv", "output_RSVA_LUGANO_3.csv",
  "output_RSVA_LUGANO_5.csv", "output_RSVA_LUGANO_7.csv", "output_RSVA_LUGANO_10.csv",
  "output_RSVA_LUGANO_15.csv", "output_RSVA_ZH_3.csv", "output_RSVA_ZH_7.csv",
  "output_RSVA_ZH_10.csv", "output_RSVA_ZH_15.csv", "output_RSVA_ZH_GE_3.csv",
  "output_RSVA_ZH_GE_5.csv", "output_RSVA_ZH_GE_7.csv", "output_RSVA_ZH_GE_10.csv",
  "output_RSVA_ZH_GE_15.csv", "output_RSVA_ZH_GE_LUG_CHUR_3.csv", 
  "output_RSVA_ZH_GE_LUG_CHUR_5.csv", "output_RSVA_ZH_GE_LUG_CHUR_7.csv",
  "output_RSVA_ZH_GE_LUG_CHUR_10.csv", "output_RSVA_ZH_GE_LUG_CHUR_15.csv"
)

# Loop through each file and assign to variables
for (file_name in csv_files) {
  var_name <- gsub(".csv", "", file_name)
  assign(var_name, {
    df <- read.csv(file = paste0(base_path, file_name), header = TRUE, check.names = FALSE)
    df$source <- var_name
    df
  })
}
```

```r
options(repr.plot.width=(10), repr.plot.height=(30))

# Convert to variable names by stripping .csv
vars_to_bind <- gsub(".csv", "", csv_files)

# Bind all the corresponding data frames together
all_data <- bind_rows(mget(vars_to_bind))



# Grouping and calculating median & MAD for each source
binary_nmf_cross_v_results <- all_data %>%
  group_by(source, W_regularization_param) %>%
  mutate(median_squared_error_loss = median(squared_error_loss),
         mad = mad(squared_error_loss)) %>%
  ungroup()

# Finding the optimal regularization parameter for each source
# optimal_params <- binary_nmf_cross_v_results %>%
#   group_by(source) %>%
#   summarise(min_y = min(median_squared_error_loss),
#             min_x = unique(W_regularization_param[which.min(median_squared_error_loss)]),
#             mad_of_min = mad(median_squared_error_loss[which.min(median_squared_error_loss)]))


optimal_params <- binary_nmf_cross_v_results %>%
  group_by(source) %>%
  filter(W_regularization_param > 0) %>%

  mutate(
      min_y = min(median_squared_error_loss),
      min_x = unique(W_regularization_param[which.min(median_squared_error_loss)]),
      mad_of_min = mad(median_squared_error_loss, na.rm = TRUE)) %>%
  filter(median_squared_error_loss <= (min_y + mad_of_min)) %>%

mutate(opt_regularization_param= max(W_regularization_param))

# Plot
plot <- ggplot(binary_nmf_cross_v_results, aes(x = (W_regularization_param), y = (median_squared_error_loss))) +
  geom_line() +
 # geom_errorbar(aes(ymin = log10(median_squared_error_loss - mad), 
   #                 ymax = log10(median_squared_error_loss + mad)), col = "red") +
  geom_vline(data = optimal_params, aes(xintercept = log10(opt_regularization_param)), col = "blue", linetype = "dashed") +
  #geom_hline(data = optimal_params, aes(yintercept = log10(min_y + mad_of_min)), col = "darkgreen", linetype = "dashed") +
  #geom_hline(data = optimal_params, aes(yintercept = log10(min_y - mad_of_min)), col = "darkgreen", linetype = "dashed") +
  labs(title = "Binarization Parameter Selection using CV",
       x = "Log10(W Regularization Parameter)",
       y = "Log10(Median of Squared Error Loss)") +
    geom_text(aes(label = round(W_regularization_param, 4), y = median_squared_error_loss + 0.1), 
            hjust = -0.5, size = 2.5, angle=90) +
  facet_wrap(~ source, scales = "free", ncol = 3, nrow = 10) +
#ylim(-5,1)+
  theme_minimal() +
  theme(
    plot.title = element_text(size = 10),
    axis.title.x = element_text(size = 18),
    axis.title.y = element_text(size = 18),
    axis.text.x = element_text(size = 14),
    axis.text.y = element_text(size = 14),
    strip.text = element_text(size = 16) # Facet labels
  )
print(plot)

ggsave("binarization_parameter_selection_2024_2025_RSVA.pdf", width = 20, height = 40, units = "in", dpi = 300)
```

```r
AA <- combined_mut_freq_ZH_GE_LUG_CHUR_3


AUPRC_distance_matrix <- matrix(data = NA, nrow = 3, ncol = 3)
rownames(AUPRC_distance_matrix) <-
  c("A.D.1","A.D.3","A.D.5.2") #c("Variant1", "Variant2", "Variant3")
colnames(AUPRC_distance_matrix) <- c(1:3)
for (r in (1:3)) {
  for (c in (1:3)) {
    predict <- prediction(w[, c],
                          reference(variant = c("A.D.1","A.D.3","A.D.5.2")[r],
                                        data_matrix = AA, 
                                        definition_matrix = definition_matrix))
    
    prec_rec <- performance(predict, 'aucpr')
    prec_rec_value <-  prec_rec@y.values[[1]]
    
    AUPRC_distance_matrix[r, c] <- prec_rec_value
    
  }
}
assignment <- solve_LSAP(AUPRC_distance_matrix, maximum = TRUE) # true -> predicted


#Binarization at 0.5 threshold
binary_w <- w
binary_w[w >= 0.5] <- 1
binary_w[w < 0.5] <- 0


print("A.D.1")
sum(abs(as.vector(unlist((reference(variant = "A.D.1",data_matrix = AA, definition_matrix =definition_matrix))))-binary_w[,assignment[1]]))
print("A.D.3")
sum(abs(as.vector(unlist((reference(variant = "A.D.3",data_matrix = AA, definition_matrix =definition_matrix))))-binary_w[,assignment[2]]))
print("A.D.5.2")
sum(abs(as.vector(unlist((reference(variant = "A.D.5.2",data_matrix = AA, definition_matrix =definition_matrix))))-binary_w[,assignment[3]]))



known_variants <- c("A.D.1","A.D.3","A.D.5.2")


# Define Jaccard similarity function:
      jaccard_similarity <- function (x, y) {
        A_B = sum(x == 1 & y == 1)
        A = sum(x == 1 & y == 0)
        B = sum(x == 0 & y == 1)
        
        return (A_B / (A + B + A_B))
      }
      
      
jaccard_similarity(as.vector((unlist((reference(variant = "A.D.1",data_matrix = AA, definition_matrix =definition_matrix))))), as.vector(unlist(binary_w[, assignment[1]])))
jaccard_similarity(as.vector((unlist((reference(variant = "A.D.3",data_matrix = AA, definition_matrix =definition_matrix))))), as.vector(unlist(binary_w[, assignment[2]])))
jaccard_similarity(as.vector((unlist((reference(variant = "A.D.5.2",data_matrix = AA, definition_matrix =definition_matrix))))), as.vector(unlist(binary_w[, assignment[3]])))



find_optimal_threshold <-
  function(predicted_vector,
           known_variant,
           data_matrix_obs_freq) {
    pred <- prediction(predicted_vector, unlist((reference(variant = known_variant, data_matrix = AA, definition_matrix = definition_matrix))))
    perf <- performance(pred, "prec", "rec")
    cutoffs <-
      data.frame(
        cut = perf@alpha.values[[1]],
        prec = perf@x.values[[1]],
        rec = perf@y.values[[1]]
      )
    precision <-  cutoffs[-c(1), 2]
    recall <- cutoffs[-c(1), 3]
    cut <- cutoffs[-c(1), 1]
    
    print(paste(max(
      fscore <-
        (2 * precision * recall) / (precision + recall), na.rm = TRUE
    ), known_variant, sep = " "))
    
    threshold <- cut[which.max(fscore)]
    
    print(precision[which.max(fscore)])
    print(recall[which.max(fscore)])
    return(threshold)
  }



threshold1 <- # predicted 1st variant
  find_optimal_threshold(
    predicted_vector = w[, (assignment[1])],
    known_variant = known_variants[1],
    data_matrix_obs_freq = AA_simulated
  )
threshold2 <-
  find_optimal_threshold(
    predicted_vector = w[, (assignment[2])],
    known_variant = known_variants[2],
    data_matrix_obs_freq = AA_simulated
  )
threshold3 <-
  find_optimal_threshold(
    predicted_vector = w[, (assignment[3])],
    known_variant = known_variants[3],
    data_matrix_obs_freq = AA_simulated
  )

thresholds <- c(threshold1, threshold2, threshold3)



vline_data <- data.frame(Variant = c("A.D.1","A.D.3","A.D.5.2"),
                         xintercept = thresholds)

#Reorder columns for the plotting:
w <- w[,c(assignment[1], assignment[2], assignment[3])]
colnames(w) <- c("A.D.1","A.D.3","A.D.5.2") # Plotting in the order of true variants
plot <-
  ggplot(data = gather(as.data.frame(w[,c(assignment[1], assignment[2], assignment[3])]), key = "Variant", value = "Mut_weights")) +
  geom_histogram(aes(x = Mut_weights, fill = Variant), bins = 50) +
  labs(x = "weight group", y = "counts", title = "Histogram of predicted weights ") +
  theme(
    axis.title = element_text(size = 18),
    axis.text = element_text(size = 18),
    plot.title = element_text(size = 18)
  ) +
  facet_wrap( ~ Variant, scales = "free_y") +
  geom_vline(
    data = vline_data,
    aes(xintercept = xintercept),
    color = "red",
    linetype = "dashed"
  )+
  scale_x_continuous(breaks = seq(0, 2, by = 0.1))

print(plot)

AUPRC_distance_matrix

```

Write predictions into a fasta file:

```r
# Function to apply mutation
mutate_sequence <- function(ref_sequence, mutation) {
    # extract the position of the mutation
    pos <- as.numeric(regmatches(mutation, gregexpr("[0-9]+", mutation))[[1]])
    # extract the mutated nucleotide (alternative variant)
    new_nuc <- (regmatches(mutation, gregexpr("[A-Z]+", mutation))[[1]])[2]
    
    seq_chars <- strsplit(as.character(ref_sequence), "")[[1]]
    # replace the reference by mutated (alternative) nucleotide
    seq_chars[pos] <- new_nuc
    return(DNAString(paste(seq_chars, collapse = "")))
}

seq_names <- colnames(predicted_W)
sequences <- c()

for (col in colnames(predicted_W)){
    new_seq <- rsva_reference

    
    for (i in rownames(predicted_W[predicted_W[, col]==1,])){

        
         new_seq <- mutate_sequence(new_seq, i)
    }
    sequences <- c(sequences, new_seq)
    
}
# Example: Extract sequences and names from a data frame
seq_set <- DNAStringSet(sequences)  # Replace with actual column name
names(seq_set) <- seq_names  # Replace with actual column name

# Write to FASTA
writeXStringSet(seq_set, "results/RSVA_2024_2025/predicted_rsv_sequences.fasta")

```

##### Deconvolve known variants (ground truth) -> compare the relative abundances with the predicted relative abundances

```r
# Definitions of the variants

W_pn_unordered <- definition_matrix[rowSums(definition_matrix[, c("A.D.1", "A.D.3", "A.D.5.2")]) > 0,
                        c("A.D.1", "A.D.3", "A.D.5.2")]

indices <- match(rownames(AA), rownames(W_pn_unordered))

# By subsetting W_pn using these indices, the rows of W_pn will be reordered to align with the row order of AA.
# In definitions I don't include those mutations that are not in the data matrix
W_pn <- W_pn_unordered[indices, ] 
rownames(W_pn) = rownames(AA)

W_pn[is.na(W_pn)] <- 0


options(repr.plot.width=7, repr.plot.height=(4))

decomposition_masked <-
  nnmf(
    as.matrix(AA),
    k = 3,
    loss = 'mse',
    method = "lee",
    check.k = FALSE,
    
    init = list(W = as.matrix(W_pn)),
    mask = list(W = matrix(data = as.logical(rep(1)), nrow = dim(AA)[1], ncol = 3)), # now all are fixed
      
    alpha = c(0.0, 0.0, 0.0), # no regularization
    beta = c(0.0, 0.0, 0.0), 
    max.iter = 500000L,
    rel.tol = 1e-5
      
  )


H_pred_fully_masked <- as.data.frame(decomposition_masked$H)

rownames(H_pred_fully_masked) <- colnames(W_pn)
H_pred_fully_masked <- rownames_to_column(H_pred_fully_masked, var = "Variant")

H_pred_fully_masked_long <- gather(H_pred_fully_masked, key = "Date", value = "Frequency", -Variant)

#H_pred_fully_masked_long$Date <- as.Date(H_pred_fully_masked_long$Date, format = "%Y-%m-%d")

H_pred_fully_masked_long <- H_pred_fully_masked_long %>%
  group_by(Date) %>%
  mutate(Sum_Frequency = sum(Frequency))

H_pred_fully_masked_long <- H_pred_fully_masked_long %>%
  mutate(Frequency_Normalized = Frequency / Sum_Frequency)

head(H_pred_fully_masked_long)


plot <- ggplot(H_pred_fully_masked_long, aes(x = Date, y = Frequency_Normalized, color = Variant, group = Variant)) +
  geom_point(size = 3) +
  labs(x = "Date", y = "Relative abundance", 
       subtitle = "", 
       caption = "") +
  theme_minimal() +  # Use a minimal theme to get a white background
  theme(legend.position = "right",
        axis.text.x = element_text(size = 12),
        axis.text.y = element_text(size = 14),
        axis.title.x = element_text(size = 14),  
        axis.title.y = element_text(size = 14),
        panel.background = element_rect(fill = "white")) + 
  scale_color_manual(values = c("A.D.1" = "red","A.D.3" = "green",  "A.D.5.2" = "blue"))# +
#  scale_x_date(date_breaks = "1 month", date_labels = "%b %Y")
plot

```
