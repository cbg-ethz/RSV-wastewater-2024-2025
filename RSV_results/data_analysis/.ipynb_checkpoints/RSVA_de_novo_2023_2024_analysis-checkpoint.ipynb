```r
# Binary Masked Matrix Factorization (BMF), adapted from [Zhang2007].

continue <- function(objective_f_values,
                     n_iter,
                     check.iter,
                     check.interval,
                     max_iter,
                     rel_tol) {
  
  if (n_iter %% check.iter == 0){
    
    if (n_iter >= max_iter) {
      return(FALSE)
    } else 
      if (n_iter > 0) if (abs((log(max(tail(unlist(objective_f_values), check.interval)))
                               - log(min(tail(unlist(objective_f_values), check.interval))))/log(max(tail(unlist(objective_f_values), 
                                                                                                          check.interval)))) < rel_tol) { 
          # abs(max(tail(unlist(objective_f_values), 50)) - min(tail(unlist(objective_f_values), 50))) < rel_tol
        return(FALSE)
      }
    else return(TRUE)
    
  } 
  return(TRUE)
}


masked_binary_nmf <- function(A,
                       W_init,
                       masked_variants, # indices of masked columns
                       rank,
                       max_iter = 50000, # maximum number of iterations before the timeout
                       rel_tol = 1e-4, # relative tolerance
                       n_run = 100,
                       lambda_w = 0, # default unregularized NMF
                       check.iter = 100,
                       check.interval = 50
) {
  # run NMF with different initialization conditions in parallel
  multiple_runs_results <- foreach(run=1:n_run, .combine=c) %dopar%{
    # log values from each run:
    objective_f_values <- c()
    relative_change <- c()
    stopping_error <- c()
    
    n_iter = 0
    
    ## Initialization
    W <- matrix(runif(nrow(A) * rank, min = 0, max = 1), nrow = nrow(A), ncol = rank)
    H <- matrix(runif(rank * ncol(A), min = 0, max = 1), nrow = rank, ncol = ncol(A))
    W[,masked_variants] <- W_init[,masked_variants] # masked variants leave unchanged during initialization step
    print(W)
    # update H matrix
    H1 <- (t(W) %*% A)
    H2 <- ((t(W) %*% W) %*% H)
    H <- H * (H1 / H2)
    
    # update W matrix
    W1 <- 2 * (A %*% t(H)) 
    W2 <- 2 * (W %*% (H %*% t(H))) 
    W[, !masked_variants] <- (W * (W1 / W2))[, !masked_variants]
    
    # Compute initial objective function
    init_obj_f_value <- sum((A - W %*% H)^2) + lambda_w * sum((W * W - W)^2) / 2
    
    c_obj <- init_obj_f_value
    p_obj <- 1e08
    
    
    while (continue(objective_f_values, n_iter, check.iter, check.interval, max_iter, rel_tol)) { 
      W[, !masked_variants] <- pmax(W[, !masked_variants], 1e-16)
      H <- pmax(H, 1e-16)
      # current objective function value becomes the "previous", and we update the current one
      p_obj <- c_obj
      # update H matrix
      H1 <- (t(W) %*% A)
      H2 <- ((t(W) %*% W) %*% H)
      H <- H * (H1 / H2)
      
      # update W matrix
      W1 <- (A %*% t(H)) + 3 * (lambda_w * (W * W))
      W2 <- (W %*% (H %*% t(H))) + 2 * lambda_w * (W * W * W) + lambda_w * W
      W[, !masked_variants] <- (W * (W1 / W2))[, !masked_variants]
      
      
      n_iter <- n_iter + 1
      c_obj <- sum((A - W %*% H)^2) + lambda_w * sum((W * W - W)^2) / 2
      
      # Update current objective function value
      objective_f_values <- c(objective_f_values, c_obj)
      relative_change <- c(relative_change, abs(p_obj - c_obj)/mean(p_obj,c_obj))
      stopping_error<- c(stopping_error, abs(diff(log(range(tail(unlist(objective_f_values), check.interval))))/log(max(tail(unlist(objective_f_values), check.interval)))))
    }
    single_run_results <- list(
      W = W,
      H = H,
      n_iter = n_iter,
      objective_f_values = objective_f_values,
      relative_change = relative_change,
      stopping_error = stopping_error)
    single_run_results
    
  }
  # Keep fitted factorization with the lowest objective function value
  
  objectives_vec <- c()
  for (i in seq_along(multiple_runs_results)){
    if(names(multiple_runs_results[i])=="objective_f_values"){
      
      objectives_vec <- c(objectives_vec, tail(multiple_runs_results[[i]], 1))
    }
  }
  run_of_min_obj <- which.min(objectives_vec)
  result <- list(W = multiple_runs_results[[(run_of_min_obj-1) * 6 + 1]], # 6=length of single run results
                 H = multiple_runs_results[[(run_of_min_obj-1) * 6 + 2]],
                 n_iter=multiple_runs_results[[(run_of_min_obj-1) * 6 + 3]],
                 objective_f_values=multiple_runs_results[[(run_of_min_obj-1) * 6 + 4]],
                 relative_change=multiple_runs_results[[(run_of_min_obj-1) * 6 + 5]],
                 stopping_error=multiple_runs_results[[(run_of_min_obj-1)* 6 + 6]])
  
  
  return(result)
  
}
```

```r
AA <- combined_mut_freq_ZH_GE_smoothed_10


AUPRC_distance_matrix <- matrix(data = NA, nrow = 3, ncol = 3)
rownames(AUPRC_distance_matrix) <-
  c("A.D.1","A.D.3","A.D.5.2") #c("Variant1", "Variant2", "Variant3")
colnames(AUPRC_distance_matrix) <- c(1:3)
for (r in (1:3)) {
  for (c in (1:3)) {
    predict <- prediction(w[, c],
                          reference(variant = c("A.D.1","A.D.3","A.D.5.2")[r],
                                        data_matrix = AA, 
                                        definition_matrix = definition_matrix))
    
    prec_rec <- performance(predict, 'aucpr')
    prec_rec_value <-  prec_rec@y.values[[1]]
    
    AUPRC_distance_matrix[r, c] <- prec_rec_value
    
  }
}
assignment <- solve_LSAP(AUPRC_distance_matrix, maximum = TRUE) # true -> predicted


#Binarization at 0.5 threshold
binary_w <- w
binary_w[w >= 0.5] <- 1
binary_w[w < 0.5] <- 0


print("A.D.1")
sum(abs(as.vector(unlist((reference(variant = "A.D.1",data_matrix = AA, definition_matrix =definition_matrix))))-binary_w[,assignment[1]]))
print("A.D.3")
sum(abs(as.vector(unlist((reference(variant = "A.D.3",data_matrix = AA, definition_matrix =definition_matrix))))-binary_w[,assignment[2]]))
print("A.D.5.2")
sum(abs(as.vector(unlist((reference(variant = "A.D.5.2",data_matrix = AA, definition_matrix =definition_matrix))))-binary_w[,assignment[3]]))



known_variants <- c("A.D.1","A.D.3","A.D.5.2")


# Define Jaccard similarity function:
      jaccard_similarity <- function (x, y) {
        A_B = sum(x == 1 & y == 1)
        A = sum(x == 1 & y == 0)
        B = sum(x == 0 & y == 1)
        
        return (A_B / (A + B + A_B))
      }
      
      
jaccard_similarity(as.vector((unlist((reference(variant = "A.D.1",data_matrix = AA, definition_matrix =definition_matrix))))), as.vector(unlist(binary_w[, assignment[1]])))
jaccard_similarity(as.vector((unlist((reference(variant = "A.D.3",data_matrix = AA, definition_matrix =definition_matrix))))), as.vector(unlist(binary_w[, assignment[2]])))
jaccard_similarity(as.vector((unlist((reference(variant = "A.D.5.2",data_matrix = AA, definition_matrix =definition_matrix))))), as.vector(unlist(binary_w[, assignment[3]])))



find_optimal_threshold <-
  function(predicted_vector,
           known_variant,
           data_matrix_obs_freq) {
    pred <- prediction(predicted_vector, unlist((reference(variant = known_variant, data_matrix = AA, definition_matrix = definition_matrix))))
    perf <- performance(pred, "prec", "rec")
    cutoffs <-
      data.frame(
        cut = perf@alpha.values[[1]],
        prec = perf@x.values[[1]],
        rec = perf@y.values[[1]]
      )
    precision <-  cutoffs[-c(1), 2]
    recall <- cutoffs[-c(1), 3]
    cut <- cutoffs[-c(1), 1]
    
    print(paste(max(
      fscore <-
        (2 * precision * recall) / (precision + recall), na.rm = TRUE
    ), known_variant, sep = " "))
    
    threshold <- cut[which.max(fscore)]
    
    print(precision[which.max(fscore)])
    print(recall[which.max(fscore)])
    return(threshold)
  }



threshold1 <- # predicted 1st variant
  find_optimal_threshold(
    predicted_vector = w[, (assignment[1])],
    known_variant = known_variants[1],
    data_matrix_obs_freq = AA_simulated
  )
threshold2 <-
  find_optimal_threshold(
    predicted_vector = w[, (assignment[2])],
    known_variant = known_variants[2],
    data_matrix_obs_freq = AA_simulated
  )
threshold3 <-
  find_optimal_threshold(
    predicted_vector = w[, (assignment[3])],
    known_variant = known_variants[3],
    data_matrix_obs_freq = AA_simulated
  )

thresholds <- c(threshold1, threshold2, threshold3)



vline_data <- data.frame(Variant = c("A.D.1","A.D.3","A.D.5.2"),
                         xintercept = thresholds)

#Reorder columns for the plotting:
w <- w[,c(assignment[1], assignment[2], assignment[3])]
colnames(w) <- c("A.D.1","A.D.3","A.D.5.2") # Plotting in the order of true variants
plot <-
  ggplot(data = gather(as.data.frame(w[,c(assignment[1], assignment[2], assignment[3])]), key = "Variant", value = "Mut_weights")) +
  geom_histogram(aes(x = Mut_weights, fill = Variant), bins = 50) +
  labs(x = "weight group", y = "counts", title = "Histogram of predicted weights ") +
  theme(
    axis.title = element_text(size = 18),
    axis.text = element_text(size = 18),
    plot.title = element_text(size = 18)
  ) +
  facet_wrap( ~ Variant, scales = "free_y") +
  geom_vline(
    data = vline_data,
    aes(xintercept = xintercept),
    color = "red",
    linetype = "dashed"
  )+
  scale_x_continuous(breaks = seq(0, 2, by = 0.1))

print(plot)

AUPRC_distance_matrix

```

Write predictions into a fasta file:

```r
# Function to apply mutation
mutate_sequence <- function(ref_sequence, mutation) {
    # extract the position of the mutation
    pos <- as.numeric(regmatches(mutation, gregexpr("[0-9]+", mutation))[[1]])
    # extract the mutated nucleotide (alternative variant)
    new_nuc <- (regmatches(mutation, gregexpr("[A-Z]+", mutation))[[1]])[2]
    
    seq_chars <- strsplit(as.character(ref_sequence), "")[[1]]
    # replace the reference by mutated (alternative) nucleotide
    seq_chars[pos] <- new_nuc
    return(DNAString(paste(seq_chars, collapse = "")))
}

seq_names <- colnames(predicted_W)
sequences <- c()

for (col in colnames(predicted_W)){
    new_seq <- rsva_reference

    
    for (i in rownames(predicted_W[predicted_W[, col]==1,])){

        
         new_seq <- mutate_sequence(new_seq, i)
    }
    sequences <- c(sequences, new_seq)
    
}
# Example: Extract sequences and names from a data frame
seq_set <- DNAStringSet(sequences)  # Replace with actual column name
names(seq_set) <- seq_names  # Replace with actual column name

# Write to FASTA
writeXStringSet(seq_set, "results/RSVA_2023_2024/predicted_rsv_sequences.fasta")


# #initialize as a reference sequence and add predicted mutations
# predicted_sequence_1 <- rsva_reference
# for (i in rownames(predicted_W[predicted_W["pred_A.D.1"]==1,])){
    
#     predicted_sequence_1 <- mutate_sequence(predicted_sequence_1, i)

#     }
# print(predicted_sequence_1)

# # Ensure the sequence is wrapped in a DNAStringSet
# predicted_sequence_1 <- DNAStringSet(predicted_sequence_1)


# writeXStringSet(predicted_sequence_1, "results/RSVA_2023_2024/pred_A.D.1_rsv_a_2023_2024_ZH.fasta")
```
